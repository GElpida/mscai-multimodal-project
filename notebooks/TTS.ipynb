```python
import edge_tts
import asyncio
import nest_asyncio
import os

nest_asyncio.apply()
os.makedirs("tts_outputs", exist_ok=True)
```

```python
texts = [
    {
        "id": "statue",
        "text": "In front of you is a marble statue of Aphrodite from the Classical period."
    },
    {
        "id": "navigation",
        "text": "Turn slightly left. The next exhibit is two meters ahead."
    }
]
```

```python
def calm_ssml(text):
    return f"""
<speak>
    <prosody rate="-15%" pitch="0Hz">
        {text}
    </prosody>
</speak>
"""
```

```python
def expressive_ssml(text):
    return f"""
<speak>
    <prosody rate="-5%" pitch="+10Hz">
        In front of you...
        <break time="300ms"/>
        {text}
        <break time="400ms"/>
        Take a moment to notice the details.
    </prosody>
</speak>
"""
```

```python
async def generate():
    for item in texts:

        # Calm
        calm = edge_tts.Communicate(
            calm_ssml(item["text"]),
            voice="en-GB-SoniaNeural"
        )
        await calm.save(f"tts_outputs/{item['id']}_calm.mp3")

        # Expressive
        expressive = edge_tts.Communicate(
            expressive_ssml(item["text"]),
            voice="en-GB-LibbyNeural"
        )
        await expressive.save(f"tts_outputs/{item['id']}_expressive.mp3")

        print("Done:", item["id"])

await generate()
```

```python
import edge_tts
import nest_asyncio
import os

nest_asyncio.apply()
os.makedirs("tts_outputs", exist_ok=True)

texts = [
    {
        "id": "statue",
        "text": "In front of you is a life-sized marble statue of a woman. The surface is smooth and cool, shaped with gentle curves that resemble the human body. Her posture is relaxed, with her weight resting slightly on one leg, giving a sense of balance and calm. Although you cannot see her face, it is carefully formed, with soft features and a peaceful expression. The statue feels still, but also full of quiet movement, as if she could shift her weight at any moment."
    },
    {
        "id": "painting",
        "text": "You are standing in front of a large painting. The scene is arranged horizontally, with figures placed across the space from left to right. At the center, there is a strong presence that draws attention, surrounded by smaller elements that feel less dominant. The composition creates a sense of depth, as if some parts are closer to you while others are further away. The overall feeling is calm but slightly dramatic, as if something important has just happened or is about to happen."
    }
]

styles = {
    "calm": {
        "voice": "en-GB-SoniaNeural",
        "rate": "-5%",
        "pitch": "-2Hz",
        "volume": "+0%",
    },

    "engaging": {
     "voice": "en-US-JennyNeural",
     "rate": "-2%",
     "pitch": "+4Hz",
     "volume": "+3%",
}
}

def make_expressive_text(text):
    return (
        "Take a moment. "
        + text.replace(".", "... ")
        + "Notice the details carefully."
    )

async def generate():
    for item in texts:
        for style_name, style in styles.items():

            if style_name == "expressive":
                final_text = make_expressive_text(item["text"])
            else:
                final_text = item["text"]

            output_path = f"tts_outputs/{item['id']}_{style_name}.mp3"

            communicate = edge_tts.Communicate(
                text=final_text,
                voice=style["voice"],
                rate=style["rate"],
                pitch=style["pitch"],
                volume=style["volume"]
            )

            await communicate.save(output_path)
            print("Saved:", output_path)

await generate()
```
